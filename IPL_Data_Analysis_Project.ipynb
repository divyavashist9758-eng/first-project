#Most Successful Teams


```python
wins=matches['winner'].value_counts()
wins.plot(kind='bar', color='pink',figsize=(10,5))
plt.title('Most Successful IPL Teams')
plt.xlabel('Team')
plt.ylabel('Wins')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

# Top Run Scores

```python
top_batsmen = deliveries.groupby('batter')['batsman_runs'].sum().sort_values(ascending=False).head(10)
top_batsmen.plot(kind='bar',color='green',figsize=(10,5))
plt.title('üî•Top 10 Batsmen by Runs')
plt.ylabel('Runs')
plt.xlabel('Batsman')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

SO Here Virat Kohli is the Top run scorer with over 6000 runs.

#Top Wicket Takers

```python
wickets= deliveries[deliveries['dismissal_kind'].notnull()]
top_bowlers=wickets['bowler'].value_counts().head(10)
top_bowlers.plot(kind='bar',color='red', figsize=(10,5))
plt.ylabel("Wickets")
plt.xlabel("Bowlers")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

#Toss Winner=Match Winner Analysis

```python
same_winner=matches[matches['toss_winner']==matches['winner']]
percentage=(len(same_winner)/len(matches))*100
f"Toss winner also won the match in { percentage:.2f}% of the matches."
```

```python
same_winner=matches[matches['toss_winner']==matches['winner']]
toss_win_and_match_win=len(same_winner)
toss_win_but_lost_match=len(matches)-toss_win_and_match_win
labels=['Toss Winner=Match Winner','Toss Winner != Match Winner']
sizes=[toss_win_and_match_win,toss_win_but_lost_match]
colors=['#4CAF50','#F44336']
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, colors=colors, autopct='%.2f%%',startangle=90,shadow=True)
plt.title('TossWinner Also Match Winner (Pie Chart)')
plt.axis('equal')
plt.show()
```
